ให้นิสิตสร้างคอลัมน์เพิ่ม โดยคอลัมน์ที่เพิ่มมานี้ จะเก็บข้อมูลของตัวอักษรตัวแรกที่ปรากฏในคอลัมน์ Cabin เช่น  
- หากในคอลัมน์ Cabin เป็นค่า "C123" ในคอลัมน์ที่สร้างจะบันทึก "C"  
- หากในคอลัมน์ Cabin เป็นค่า "C23 C25 C27" ในคอลัมน์ที่สร้างจะบันทึก "C"
- หากในคอลัมน์ Cabin เป็นค่า "B58 B60 C2" ในคอลัมน์ที่สร้างจะบันทึก "B"
- หากในคอลัมน์ Cabin เป็นค่า "F2" ในคอลัมน์ที่สร้างจะบันทึก "F"
- หากในคอลัมน์ Cabin เป็นค่า NaN ในคอลัมน์ที่สร้างจะบันทึก NaN

จากนั้นให้นิสิตเขียนโค้ดเพื่อตอบคำถามในแต่ละข้อ

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/ThammakornS/ProgStat/main/dataset/titanic.csv")
df.head()
```

นิสิตแสดงวิธีสร้างคอลัมน์

```python
#your code
df['cabin_letter'] = df['Cabin'].str[0]
```

1. ในแต่ละตัวอักษร Cabin มีจำนวนผู้โดยสารกี่คน และมีจำนวน NaN เท่าไหร่

```python
#your code
df['cabin_letter'].value_counts(dropna=False)
```

2. ในแต่ละตัวอักษร Cabin มีจำนวนผู้โดยสารที่รอดและเสียชีวิตอย่างละเท่าไหร่ cabin ใดมีจำนวนผู้เสียชีวิตมากที่สุด

```python
#your code
# 1. สร้างตารางแจกแจงจำนวนผู้รอด (1) และเสียชีวิต (0) ตามตัวอักษร Cabin
result = pd.crosstab(df['cabin_letter'], df['Survived'])
result
```

```python
result = df.groupby(['cabin_letter', 'Survived']).size().unstack(fill_value=0)
result
```

```python
# 2. หา Cabin ที่มีผู้เสียชีวิต (col 0) มากที่สุด
# เลือกคอลัมน์ 0 (เสียชีวิต) แล้วหา index ที่มีค่ามากที่สุด
most_deaths = result[0].idxmax()
count_deaths = result[0].max()

print(f"Cabin ที่มีผู้เสียชีวิตมากที่สุดคือ: {most_deaths} (จำนวน {count_deaths} คน)")
```

3. cabin ใดมีค่า สัดส่วน จำนวนผู้รอดชีวิตหารด้วยจำนวนผู้เสียชีวิต สูงสุด

```python
#your code
# คำนวณสัดส่วน (Ratio) = รอด / เสียชีวิต
survival_ratio = result[1] / result[0]
```

```python
survival_ratio
```

```python
max_cabin = survival_ratio.idxmax()
max_value = survival_ratio.max()

print(f"Cabin ที่มีสัดส่วน รอด/เสียชีวิต สูงที่สุดคือ: {max_cabin} (สัดส่วน {max_value:.2f})")
```

4. cabin ใดมีค่าเฉลี่ยราคาตั๋ว (Fare) สูงสุด

```python
#your code
# 1. หาค่าเฉลี่ยราคาตั๋วในแต่ละ Cabin
avg_fare = df.groupby('cabin_letter')['Fare'].mean()
avg_fare
```

```python
# 2. หา Cabin ที่มีราคาเฉลี่ยสูงสุด
max_cabin_fare = avg_fare.idxmax()
max_val_fare = avg_fare.max()

print(f"Cabin ที่มีค่าเฉลี่ยราคาตั๋วสูงสุดคือ: {max_cabin_fare} (เฉลี่ย {max_val_fare:.2f})")
```

5. จงแสดงค่าเฉลี่ยราคาตั๋ว (Fare) ของผู้รอดชีวิต และผู้เสียชีวิต ในแต่ละ cabin หากเปรียบเทียบราคาเฉลี่ยของผู้ที่รอดและเสียชีวิตจะพบว่า

```python
# your code
fare_comparison = df.groupby(['cabin_letter', 'Survived'])['Fare'].mean().unstack()
fare_comparison
```
